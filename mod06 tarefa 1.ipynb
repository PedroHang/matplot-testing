```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.drop_duplicates().shape
df = pd.DataFrame(sinasc)
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig1 = plt.figure()
fig1.suptitle('APGAR 1')   
counts1 = df['APGAR1'].value_counts()
ax1 = fig1.add_subplot(1,1,1)
plt.bar(counts1.index, counts1)
```

```python
fig2 = plt.figure()
fig2.suptitle('APGAR 5')   
counts2 = df['APGAR5'].value_counts()
ax2 = fig2.add_subplot(1,1,1)
plt.bar(counts2.index, counts2)
```

```python
fig3 = plt.figure()
fig3.suptitle('SEXO')   
counts3 = df['SEXO'].value_counts()
ax3 = fig3.add_subplot(1,1,1)
plt.bar(counts3.index, counts3)
```

```python
fig4 = plt.figure()
fig4.suptitle('GESTACAO')   
counts4 = df['GESTACAO'].value_counts()
ax4 = fig4.add_subplot(1,1,1)

plt.xticks(rotation=90)
plt.subplots_adjust(bottom=0.3)
plt.bar(counts4.index, counts4)
```

```python
fig5 = plt.figure()
fig5.suptitle('ESTCIVMAE')   
counts5 = df['ESTCIVMAE'].value_counts()
ax5 = fig5.add_subplot(1,1,1)


plt.bar(counts5.index, counts5)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
fig6 = plt.figure()
fig6.suptitle('APGAR 5 MÉDIO POR CONSULTA:') 
pre_filtered1 = df.loc[df['CONSULTAS'] == 1, 'APGAR5']
filtered1 = pre_filtered1.mean()

pre_filtered2 = df.loc[df['CONSULTAS'] == 2, 'APGAR5']
filtered2 = pre_filtered2.mean()

averages = [filtered1, filtered2]
labels = ['0 Consultas', '1 a 3 Consultas']
colors = ['red', 'blue']

ax6 = fig6.add_subplot(1,1,1)
plt.bar(labels, averages, color=colors)
```

<h3>Nota-se pouca correlação entre o número de consultas e o valor APGAR 5 ao nascer</h3>
<h4> No entanto não podemos desconsiderar os benefícios de consultas pré-natais para o desenvolvimento saudável
da criança.</h4>
